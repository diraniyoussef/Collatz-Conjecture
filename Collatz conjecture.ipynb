```python
def getSelectedTuple(sym_index):
    l = []
    for i in range(len(d)):
        if i == sym_index:
            l.append((d[i],1))
            continue
        l.append((d[i],0))
    return tuple(l)
```

```python
def addtoList(binary_str, sym_index, L):
    if sym_index == -1:
        digit = 1
    else:
        digit = d[sym_index]
    for i in range(len(binary_str)):
        if binary_str[-1-i] == "1":
            L[-1-i] += digit
    return L
```

```python
def fixList(l, sym_index=-1, sym_val=1):
    L = [0 for _ in range(len(l))]
    dec = convertToDecimal(l)
    #print(dec)
    if type(dec) is int:
        non_symbol_str = bin(dec)        
        L = addtoList(non_symbol_str, -1, L)
        return L
    
    original_non_symbol_num = non_symbol_num = dec.subs(getSelectedTuple(-1))
    #print(non_symbol_num)
    if sym_index != -1 and sym_val == 1:
        #print(dec.subs(getSelectedTuple(sym_index)))
        non_symbol_num = dec.subs(getSelectedTuple(sym_index)) #note that original_non_symbol_num is implicitly included within

    non_symbol_str = bin(non_symbol_num)
    #print(non_symbol_num, non_symbol_str)
    L = addtoList(non_symbol_str, -1, L)
            
    for i in range(len(d)):
        if i == sym_index:
            continue
        num = dec.subs(getSelectedTuple(i))
        binary_str = bin(num - original_non_symbol_num)
        #print(num, binary_str)
        L = addtoList(binary_str, i, L)
    
    return L
```

```python
def subListSymbol(l, sym_index, sym_val=1): #sym_val is either 1 or 0
    return fixList(l, sym_index, sym_val)
```

```python
def nextNum(l, show_trunc=False):
    if l[-1] != 1:
        print("not odd !")
        return
    i = len(l)-1
    last_truncate_index = len(l)
    while True:
        i -= 1
        if i < 0 or type(l[i]) is not int:
            if last_truncate_index != len(l):
                return truncate(l, last_truncate_index)
            else:
                print("couldn't find a non-symbol pattern")
                return l
        if l[i] == 0 and l[i+1] == 0:
            if show_trunc and last_truncate_index != len(l):
                return truncate(l, last_truncate_index)
            return zeroZero(truncate(l,i+2))#i+2 because it's  0 0 1
        if l[i] == 1 and l[i+1] == 1:
            if show_trunc and last_truncate_index != len(l):
                return truncate(l, last_truncate_index)
            return oneOne(truncate(l,i+1)) #i+1 because it's  1 1
        if l[i] == 1 and l[i+1] == 0:  
            last_truncate_index = i
            continue
```

```python
#this keeps the 1010101 pattern in its place
def isOne(l):
    for i in range(len(l)-1):
        if l[i] != 0:
            return False
    return l[i] == 1

def truncate(l, most_included_element_index): 
    length = len(l)
    L = l.copy()
    for i in range(length -1 - most_included_element_index):
        L.pop()
        #L.insert(0, 0)
    return L

def nextNum(l, show_trunc=False):
    if l[-1] != 1:
        print("not odd !")
        return
    i = len(l)-1
    last_truncate_index = len(l)
    while True:
        i -= 1
        if i < 0 or type(l[i]) is not int:
            print("couldn't find a non-symbol pattern")
            return l
        if l[i] == 0 and l[i+1] == 0:
            if show_trunc and last_truncate_index != len(l):
                return truncate(l, last_truncate_index)
            l1 = truncate(l,i+2)
            if isOne(l1):
                return l
            zZ = zeroZero(l1)
            if len(l) > i+3:
                zZ.extend(l[i+3:])#i+2 because it's  0 0 1
            return zZ
        if l[i] == 1 and l[i+1] == 1:
            if show_trunc and last_truncate_index != len(l):
                return truncate(l, last_truncate_index)
            oO = oneOne(truncate(l,i+1))
            if len(l) > i+2: 
                oO.extend(l[i+2:]) #i+1 because it's  1 1
            return oO
        if l[i] == 1 and l[i+1] == 0:  
            last_truncate_index = i
            continue
```
